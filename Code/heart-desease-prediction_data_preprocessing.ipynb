```python
import pandas as pd
heart_condition = pd.read_csv(r"/kaggle/input/personal-key-indicators-of-heart-disease/2022/heart_2022_no_nans.csv")
heart_condition
```

```python
heart_condition_copy = pd.read_csv(r"/kaggle/input/personal-key-indicators-of-heart-disease/2022/heart_2022_with_nans.csv")
heart_condition_copy.head()
```

```python
heart_condition_copy.isnull().sum()
```

```python
# Find the columns with missing values
columns_with_missing_values = heart_condition_copy.columns[heart_condition_copy.isnull().any()].tolist()
columns_with_missing_values
```

```python
import numpy as np
```

```python
numeric_data = heart_condition_copy.select_dtypes(include=[np.number])
categorical_data = heart_condition_copy.select_dtypes(exclude=[np.number])
```

```python
categorical_data
```

```python
 # Impute the missing values with the mean value of that column
for col in numeric_data:
    numeric_data[col].fillna(numeric_data[col].mean(), inplace=True)
        
```

```python
categorical_data = categorical_data.apply(lambda x: x.fillna(x.value_counts().index[0]))
categorical_data
```

```python
heart_condition_data = pd.concat([categorical_data, numeric_data], axis=1)
```

```python
heart_condition_data.head()
```
