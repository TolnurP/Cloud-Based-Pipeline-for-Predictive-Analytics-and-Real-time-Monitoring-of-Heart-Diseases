```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('personal-key-indicators-of-heart-disease.csv')
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
# Display the first few rows of the dataframe
pd.set_option('display.max_columns', None)
df.head()
```

```python
# Get basic info about the dataframe
df.info()
```

```python
# Summary statistics for numerical columns
df.describe()
```

```python
# Check for any missing values
df.isnull().sum()
```

```python
# Check for duplicate rows
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
# Check for duplicate rows
df.duplicated().sum()
```

```python
df.to_csv('Heart_Disease_Prediction_Data_Final.csv', index=False)
```

```python
# Univariate Analysis - Numerical Features
numerical_cols = ['PhysicalHealthDays', 'MentalHealthDays', 'SleepHours', 'HeightInMeters', 'WeightInKilograms', 'BMI']
df[numerical_cols].hist(bins=15, figsize=(15, 10))
plt.show()
```

```python
# Boxplots for numerical variables to identify outliers
for col in numerical_cols:
    plt.figure(figsize=(6, 4))
    sns.boxplot(data=df, x=col)
    plt.show()
```

```python
# Univariate Analysis - Categorical Features
categorical_cols = [col for col in df.columns if df[col].dtype == 'object']
for col in categorical_cols:
    plt.figure(figsize=(10, 5))
    sns.countplot(y=col, data=df)
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Bivariate Analysis - Correlation Matrix for Numerical Features
plt.figure(figsize=(10, 8))
sns.heatmap(df[numerical_cols].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.show()
```

```python
# Distribution of target variable or key indicators (e.g., 'HadHeartAttack')
plt.figure(figsize=(8, 6))
sns.countplot(x='HadHeartAttack', data=df)
plt.title('Distribution of HadHeartAttack')
plt.show()
```

```python
# Explore the distribution of numeric features
numeric_features = df.select_dtypes(include=['float64']).columns
for feature in numeric_features:
    plt.figure(figsize=(8, 6))
    sns.histplot(df[feature], kde=True)
    plt.title(f'Distribution of {feature}')
    plt.show()
```

```python
# Explore the relationship between features and the target variable
categorical_features = df.select_dtypes(include=['object']).columns
for feature in categorical_features:
    plt.figure(figsize=(8, 6))
    sns.countplot(x=feature, hue='HadHeartAttack', data=df)
    plt.title(f'{feature} vs HadHeartAttack')
    plt.show()
```

```python
# Pairplot for a quick overview of relationships
sns.pairplot(df)
plt.show()
```
