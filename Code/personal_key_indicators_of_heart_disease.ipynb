```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
```

```python
# Load the dataset
heart_condition_data1 = pd.read_csv('personal-key-indicators-of-heart-disease.csv')
heart_condition_data1
```

```python
# Load the dataset
heart_condition_data2 = pd.read_csv('heart_attack_psotive_data.csv')
heart_condition_data2
```

```python
# Concatenate the datasets
heart_condition_data = pd.concat([heart_condition_data1, heart_condition_data2], axis=0, ignore_index=True)

# Display the concatenated dataset
print(heart_condition_data)
```

```python
from sklearn.utils import resample

# Separate majority and minority classes
majority_class = heart_condition_data[heart_condition_data.HadHeartAttack == 'No']
minority_class = heart_condition_data[heart_condition_data.HadHeartAttack == 'Yes']

upsample_size = int(majority_class.shape[0] * 0.3)

# Upsample minority class
minority_upsampled = resample(minority_class,
                              replace=True,     # sample with replacement
                              n_samples=upsample_size,    # to match majority class
                              random_state=123) # reproducible results

# Combine majority class with upsampled minority class
heart_condition_data_upsampled = pd.concat([majority_class, minority_upsampled])

# Shuffle the dataset
heart_condition_data_upsampled = heart_condition_data_upsampled.sample(frac=1, random_state=1275).reset_index(drop=True)
```

## **Explore the Data**

```python
# Display basic information about the upsampled dataset
print(heart_condition_data_upsampled.info())

# Display summary statistics
print(heart_condition_data_upsampled.describe())

# Check the distribution of the target variable
print(heart_condition_data_upsampled['HadHeartAttack'].value_counts())
```

```python
heart_condition_data_upsampled = heart_condition_data_upsampled.drop('Unnamed: 0', axis=1)
```

```python
heart_condition_data_upsampled.columns
```

```python
# Separate features and target variable
X = heart_condition_data_upsampled.drop('HadHeartAttack', axis=1)
y = heart_condition_data_upsampled['HadHeartAttack']
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=76)
```

```python
# Define the numeric columns excluding the target variable
numeric_columns = X.select_dtypes(exclude=['object']).columns

# Define the categorical columns excluding the target variable
categorical_columns = X.select_dtypes(include=['object']).columns
print("Numeric columns:", numeric_columns)
print("categorical_columns", categorical_columns)
```

```python
# Define the preprocessing steps using a ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_columns),
        ('cat', OneHotEncoder(drop='first'), categorical_columns)
    ])
```

```python
# Create a new pipeline with imputation in the preprocessing steps
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', RandomForestClassifier(n_estimators=50, random_state=76))
])
```

```python
# Check for null values in the entire DataFrame
null_values = X.isnull().sum()
null_values
```

```python
# Train the model using the pipeline
pipeline.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred = pipeline.predict(X_test)
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
```

```python
# Display classification report
print('Classification Report:')
print(classification_report(y_test, y_pred))
```

```python
# Display confusion matrix
print('Confusion Matrix:')
print(confusion_matrix(y_test, y_pred))
```

```python
# Display feature importance
feature_importance = pd.DataFrame({'Feature': X.columns, 'Importance': pipeline.named_steps['model'].feature_importances_})
feature_importance = feature_importance.sort_values(by='Importance', ascending=False)
print(feature_importance)
```
